```python
# Importing the Dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Load the dataset to pandas Dataframe
# Dataset Link
# https://www.kaggle.com/datasets/arockiaselciaa/creditcardcsv
credit_card_data = pd.read_csv("creditcard.csv")
```

```python
# First 5 rows of the Dataset
credit_card_data.head()
```

```python
# Last 5 rows of the Dataset
credit_card_data.tail()
```

```python
# Dataset information
credit_card_data.info()
```

```python
# Checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Distribution of Legit Transaction and fraudulent transaction
credit_card_data['Class'].value_counts()
```

```python
# Seperating the Data for analysis
Legit = credit_card_data[credit_card_data['Class'] == 0] # pandas series datatype
Fraud = credit_card_data[credit_card_data['Class'] == 1] # pandas series datatype
```

```python
print(Legit.shape)
print(Fraud.shape)
```

```python
# Statistical measures of Data
Legit.Amount.describe()
```

```python
Fraud.Amount.describe()
```

```python
# Compare the values for both Transactions
credit_card_data.groupby('Class').mean()
```

```python
# Select Random Sample from Legit Transactions to form Uniform or normal distibution with Fraud Transactions datapoints
Legit_sample = Legit.sample(n = 492,replace = False) # Randomly select data from 284315 data points and not First 492 records
```

```python
# Concatenating Two DataFrames
credit_card_data_new =  pd.concat([Legit_sample,Fraud],axis = 0)
```

```python
# First 5 records
credit_card_data_new.head()
```

```python
# Last 5 Records
credit_card_data_new.tail()
```

```python
# Checking the Distribution of the Data
credit_card_data_new['Class'].value_counts()
```

```python
# Comapare the values of Both Classes
credit_card_data_new.groupby('Class').mean()
```

```python
# Splitting the Data into Independent(Input Feature) and Dependent(Output Feature) variable
X = credit_card_data_new.drop(columns='Class',axis = 1)
y = credit_card_data_new['Class']
```

```python
X.head()
```

```python
y.head()
```

```python
# Split the Data into Training Data and Testing Data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y,random_state=2) # Creating 4 arrays
# stratify ensures evenly distribution of 2 classes in X_train and X_test
```

```python
# Displaying the shape of the Distribution
print(X_train.shape),print(X_test.shape),print(y_train.shape),print(y_test.shape)
```

#### Model Training

```python
model = LogisticRegression(max_iter= 2000)
```

```python
# Training the Logistic Regression model with Training data
model.fit(X_train,y_train)
```

### Model Evaluation

```python
# Accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,y_train)
```

```python
# Evaluating Accuracy on training data
print("Accuracy on Training Data : ",training_data_accuracy)
```

```python
# Accuracy score on Test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,y_test)
```

```python
# Evaluating Accuracy on test data
print("Accuracy on Training Data : ",test_data_accuracy)
```
